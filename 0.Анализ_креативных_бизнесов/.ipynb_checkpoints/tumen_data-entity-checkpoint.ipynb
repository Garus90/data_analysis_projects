# Анализ юридических лиц, работающих в сфере креативных индустрий  в городе Тюмень в 2022 году

```python
import pandas as pd
import re
import math
import warnings
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_colwidth',150)
pd.set_option('display.max_columns',20)
```

```python
data1 = pd.read_excel('/Users/ruslangayfullin/Documents/datasets/tumen/companies_main_CI/tumen_companies_1.xlsx',sheet_name='report')
data2 = pd.read_excel('/Users/ruslangayfullin/Documents/datasets/tumen/companies_main_CI/tumen_companies_2.xlsx',sheet_name='report')
```

```python
data_raw = pd.concat([data1,data2]).reset_index(drop=True)
```

```python
data_raw = data_raw.drop(columns='№')
```

```python
data_acitve = data_raw.query('Статус == "Действующая"')
data_closed = data_raw.query('Статус != "Действующая"')
```

```python
warnings.filterwarnings("ignore")
data_acitve[['Наименование','Наименование полное','Адрес (место нахождения)']] = data_acitve[['Наименование','Наименование полное','Адрес (место нахождения)']].drop_duplicates()
data_acitve = data_acitve.dropna(subset=['Наименование','Наименование полное','Адрес (место нахождения)']).reset_index(drop=True)
```

```python
data_closed[['Наименование','Наименование полное','Адрес (место нахождения)']] = data_closed[['Наименование','Наименование полное','Адрес (место нахождения)']].drop_duplicates()
data_closed = data_closed.dropna(subset=['Наименование','Наименование полное','Адрес (место нахождения)']).reset_index(drop=True)
```

```python
data_raw2 = pd.concat([data_closed,data_acitve]).reset_index(drop=True)
```

```python
data_raw2['registration_year'] = pd.DatetimeIndex(data_raw2['Дата регистрации']).year
data_raw2['close_year'] = pd.DatetimeIndex(data_raw2['Дата ликвидации']).year
```

```python
print('В соответствии с анализом баз данных Спарк-Интерфакс и Тасс-Бизнес, в Тюмени зарегистрированно {0} юридических лиц, работающих в сфере креативных индустрий (имеющих ОКВЭДЫ, относящиеся к креаивным индустриям в качестве основных или смежных). Выборка сформирована в июне 2022 года по фильтрам следующих {1} видов деятельности, наиболее распорстраненные из которых:'.format(len(data_raw2),len(data_raw2['Вид деятельности/отрасль'].unique())))
data_popular = pd.DataFrame(data_raw2['Вид деятельности/отрасль'].value_counts().reset_index().head(10).rename(columns={'index':'вид_деятельности','Вид деятельности/отрасль':'количество организаций'}))

plt.figure(figsize=(10, 20))
plt.grid(True)
sns.set(font_scale = 1)
ax = sns.barplot(y='вид_деятельности',x='количество организаций',data = data_popular,label = 'Total',);
ax.set_yticklabels(data_popular['вид_деятельности'],fontsize = 20)

plt.show()
```

## Анализ закрывшихся по годам предприятий

```python
data_closed_pivot = (
    (data_closed
 .query('close_year == 2012')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2012'})
 .sort_values(by='2012',ascending=False))
    
    .merge((data_closed
 .query('close_year == 2013')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2013'})
 .sort_values(by='2013',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2014')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2014'})
 .sort_values(by='2014',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2015')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2015'})
 .sort_values(by='2015',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2016')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2016'})
 .sort_values(by='2016',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2017')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2017'})
 .sort_values(by='2017',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2018')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2018'})
 .sort_values(by='2018',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2019')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2019'})
 .sort_values(by='2019',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2020')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2020'})
 .sort_values(by='2020',ascending=False)),
    on='index'
    ,how='outer')
    
    .merge((data_closed
 .query('close_year == 2021')['okved']
 .value_counts()
 .reset_index().rename(columns={'okved':'2021'})
 .sort_values(by='2021',ascending=False)),
    on='index'
    ,how='outer')
    
    
).fillna(0)
```
